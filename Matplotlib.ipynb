```python
# SUBPLOT: yan yana 2 tane grafik çizdirmek istersem eğer 

plt.subplot(1,2,1)#1 sıra olacak 2 kolon olacak şu anda 2. grafiği çiziyorum
plt.plot(numpyDızısı1,numpyDızısı2,"r--")
plt.subplot(1,2,2)#1 sıra olacak 2 kolon olacak şu anda 2. grafiği çiziyorum
plt.plot(numpyDızısı2,numpyDızısı1,"g--")
```

```python
#figure: boş grafik çiz

benımFıgure=plt.figure()
figureAxes=benımFıgure.add_axes([0.2,0.2,0.9,0.9])
figureAxes.plot(numpyDızısı1,numpyDızısı2,"g")
figureAxes.set_xlabel(" x ekseni veri ismi")
figureAxes.set_ylabel("y ekseni veri ismi")
figureAxes.set_title("grafik başlığı")
```
